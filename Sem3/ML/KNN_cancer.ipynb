```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as mlp
from sklearn.model_selection import train_test_split
from sklearn.neighbors import KNeighborsClassifier
from sklearn import metrics
```

```python
df=pd.read_csv('/content/Cancer_DS - Cancer_DS.csv')
```

```python
df.dtypes
```

```python
M=df[df.diagnosis =='M']
B=df[df.diagnosis =='B']
mlp.scatter(M.radius_mean,M.texture_mean,color='blue',label='maligant',alpha=.9)
mlp.scatter(B.radius_mean,B.texture_mean,color='red',label='bengin',alpha=.9)
mlp.xlabel('radius mean')
mlp.ylabel('texture mean')
mlp.legend()
mlp.show()
```

```python
feature=df[['radius_mean','texture_mean','perimeter_mean','area_mean','smoothness_mean','compactness_mean','concavity_mean','concave points_mean','symmetry_mean','fractal_dimension_mean','radius_se','texture_se','perimeter_se','area_se','smoothness_se','compactness_se','concavity_se','concave points_se','symmetry_se','fractal_dimension_se','radius_worst','texture_worst','perimeter_worst','area_worst','smoothness_worst','compactness_worst','concavity_worst','concave points_worst','symmetry_worst','fractal_dimension_worst']]
labels=df[['diagnosis']]
```

```python
x_train,x_test,y_train,y_test=train_test_split(feature,labels,test_size=.2,random_state=42)
```

```python
knn=KNeighborsClassifier(n_neighbors=5)
```

```python
knn.fit(x_train,y_train)
```

```python
y_pred=knn.predict(x_test)
```

```python
accuracy=metrics.accuracy_score(y_test,y_pred)
```

```python
print(accuracy)
```

```python
from sklearn.preprocessing import LabelEncoder
le=LabelEncoder()
cols=['Sex','Cabin','Embarked']
for i in cols:
  train[i]=le.fit_transform(train[i])
```

```python
x=train.drop(['PassengerId','Survived','Name','Ticket'],axis=1)
```

```python
print(x)
```

```python
y=train['Survived']
```

```python
from sklearn.naive_bayes import GaussianNB
model=GaussianNB()
```

```python
model.fit(x,y)
```

```python
predictiion=model.predict(x)
```

```python
from sklearn.metrics import accuracy_score
```

```python
accuracy=accuracy_score(y,predictiion)
```

```python
df.plot(kind="box")
plt.show()
```

```python
sns.set(style='whitegrid')
plt.figure(figsize=(10,6))
sns.histplot(data=df,x='sepal.length',kde=True,color='skyblue')
plt.xlabel('Sepal length')
plt.title('Distribution chart of sepal length')
plt.show()
```

```python
df.info(plt.scatter(df['sepal.length'], df['sepal.width']))
plt.xlabel('sepal length')
plt.ylabel('sepal width')
plt.title('scatter plot of sepal length vs width')
plt.show()
```

```python
from pandas.plotting import scatter_matrix
scatter_matrix(df,figsize=(10,10))
plt.show()
```

```python
plt.scatter(df['sepal.length'],df['sepal.width'],s=df['petal.length'],alpha=0.5)
plt.xlabel('sepal length')
plt.ylabel('sepal width')
plt.suptitle('Bubble chart of sepal length vs sepal width')
plt.show()
```

```python
sns.kdeplot(data=df,x='sepal.length',y='sepal.width',cmap='Blues',Shade=True)
plt.xlabel('sepal length')
plt.ylabel('sepal width')
plt.title('2D Density chart of sepal length and width')
plt.show()
```

```python
plt.figure(figsize=(10,6))
pd.plotting.parallel_coordinates(df,'variety',colormap='viridis')
plt.title=('parallel plot')
plt.show()
```

```python
reference=df['sepal.length'].mean()
df['deviation']=df['sepal.length'] - reference
plt.figure(figsize=(10,6))
plt.bar(df['variety'],df['deviation'],color='skyblue')
plt.xlabel('variety')
plt.ylabel('Deviation from mean sepal length')
plt.suptitle('Deviation Chart')
plt.show()
```
